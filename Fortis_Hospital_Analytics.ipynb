```python
import pandas as pd
import numpy as np
import random

# Set random seed for reproducibility
np.random.seed(42)
n_rows = 1000

# Generate synthetic dataset for Fortis Hospital
fortis_data = {
    'Patient_ID': [f'FOR-{10000 + i}' for i in range(n_rows)],
    'Age': np.random.randint(1, 90, size=n_rows),
    'Gender': np.random.choice(['Male', 'Female'], size=n_rows),
    'Department': np.random.choice(['Cardiology', 'Neurology', 'Oncology', 'Pediatrics', 'Orthopedics'], size=n_rows),
    'Doctor_Name': np.random.choice(['Dr. Sharma', 'Dr. Verma', 'Dr. Gupta', 'Dr. Reddy', 'Dr. Mishra'], size=n_rows),
    'Admission_Type': np.random.choice(['Emergency', 'Elective', 'Urgent'], size=n_rows, p=[0.4, 0.4, 0.2]),
    'Length_of_Stay_Days': np.random.randint(1, 20, size=n_rows),
    'Total_Bill_INR': np.random.randint(15000, 500000, size=n_rows),
    'Insurance_Provider': np.random.choice(['Star Health', 'HDFC Ergo', 'Max Bupa', 'Self-Pay'], size=n_rows),
    'Readmitted_30_Days': np.random.choice(['Yes', 'No'], size=n_rows, p=[0.15, 0.85])
}

# Convert dictionary to pandas DataFrame
df = pd.DataFrame(fortis_data)

# Introduce missing values in 'Total_Bill_INR' for data cleaning practice
df.loc[df.sample(frac=0.05).index, 'Total_Bill_INR'] = np.nan

# Display the dataframe
df
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Handle missing values by filling them with the median bill of each department
df['Total_Bill_INR'] = df['Total_Bill_INR'].fillna(df.groupby('Department')['Total_Bill_INR'].transform('median'))

# Re-check for missing values to confirm cleanup
df.isnull().sum()
```

```python
# Save the clean data for SQL and Power BI
df.to_csv("fortis_hospital_clean.csv", index=False)
print("File Saved Successfully!")
```
